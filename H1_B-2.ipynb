```python
import pandas as pd
```

```python

# Read Excel file
df = pd.read_excel('employer.xlsx')
```

```python
df.head()
```

```python

# Remove rows with missing values from the "employer name" column
df.dropna(subset=['Employer (Petitioner) Name'], inplace=True)
```

```python
df.head()
```

```python
columns_to_drop = ['Initial Approval', 'Initial Denial', 'Continuing Denial']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

```python

df = df[df['Continuing Approval'] != 0]

df.sort_values(by='Continuing Approval', ascending=False, inplace=True)
```

```python
df.head()
```

```python
df.to_excel('/content/data.xlsx', index=False)
```
